# metadata of pdf  

```python
 from pdfrw import PdfReader
 
def get_pdf_info(path):
    pdf = PdfReader(path)
    
    print(pdf.keys())
    print(pdf.Info)      
    print(pdf.Root.keys())
    print('PDF has {} pages'.format(len(pdf.pages)))
    f = open("HDFC_1.txt","a")
    f.write(f'{pdf.Info}')
    
    
 
if __name__ == '__main__':
    get_pdf_info('HDFC10.pdf')
    get_pdf_info('HDFC25.pdf')
```
